# Healthcare Payments Data (HPD) Healthcare Measures Dataset


# Overview

This dataset provides detailed insights into healthcare measures across California, focusing on three main categories:

Health Conditions: Data on the prevalence of long-term illnesses and major medical events, such as diabetes and heart failure.

Utilization: Metrics on healthcare system usage, including emergency department visits and various inpatient stays (e.g., maternity, surgical).

Demographics: Information on health coverage and demographic characteristics like age, sex, and county.
The data is organized to allow for the calculation of percentages, rates, and averages across different measures and demographics.

# Column Definitions
reporting_year: The year the data was reported.

age_band: Age group of the individuals in the dataset (e.g., 0-17).

assigned_sex_at_birth: The sex assigned to individuals at birth (Male, Female).

county_name: The county in California where the data was collected.

covered_california_region: Specific regions in California as defined by the Covered California health insurance marketplace.

payer_type: Type of health insurance payer (e.g., Commercial, Medi-Cal, Medicare).

measure_id: Identifier for the specific healthcare measure being reported.

measure_numerator: The count or sum of the specific healthcare measure (e.g., number of cases).

measure_denominator: The base population counts relevant to the measure.

measure_scaling_factor: A factor used to scale the measure, indicating its granularity.

suppression_ind: Indicates whether data suppression has been applied to protect privacy (Y for Yes, N for No).

# Key Insights

Healthcare Trends: Analyze the prevalence of certain health conditions and how they vary by region, age, and insurance type.

Utilization Patterns: Understand how different demographic groups use healthcare services like emergency departments and inpatient stays.

Demographic Disparities: Explore differences in healthcare coverage and service usage across various counties and demographic groups in California.

This dataset is valuable for healthcare analysts and policymakers seeking to understand and improve healthcare services and outcomes in California.

# Data Visualization

# Count Plot of Payer Types

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=healthcare_df, x='payer_type')
plt.title('Count of Records by Payer Type')
plt.ylabel('Count')
plt.xlabel('Payer Type')
plt.show()
```

# Distribution of Measure Numerator by Sex

```python
plt.figure(figsize=(12, 6))
sns.boxplot(data=healthcare_df, x='assigned_sex_at_birth', y='measure_numerator')
plt.title('Distribution of Measure Numerator by Sex')
plt.ylabel('Numerator of Measures')
plt.xlabel('Sex Assigned at Birth')
plt.show()
```

# Measure Scaling Factor by Age Band

```python
plt.figure(figsize=(12, 6))
sns.lineplot(data=healthcare_df, x='age_band', y='measure_scaling_factor', hue='assigned_sex_at_birth', marker='o')
plt.title('Measure Scaling Factor by Age Band and Sex')
plt.ylabel('Measure Scaling Factor')
plt.xlabel('Age Band')
plt.show()
```

# Heatmap of Measure Counts by Year and Payer Type

```python
plt.figure(figsize=(12, 8))
pivot_table = healthcare_df.pivot_table(index='reporting_year', columns='payer_type', values='measure_id', aggfunc='count')
sns.heatmap(pivot_table, annot=True, cmap='Blues')
plt.title('Heatmap of Measure Counts by Year and Payer Type')
plt.ylabel('Reporting Year')
plt.xlabel('Payer Type')
plt.show()
```

# Histogram of Measure Numerator

```python
plt.figure(figsize=(10, 6))
sns.histplot(healthcare_df['measure_numerator'], bins=30, kde=True)
plt.title('Histogram of Measure Numerator')
plt.xlabel('Measure Numerator')
plt.ylabel('Frequency')
plt.show()
```

# Scatter Plot of Measure Numerator vs Denominator

```python
plt.figure(figsize=(12, 6))
sns.scatterplot(data=healthcare_df, x='measure_denominator', y='measure_numerator', hue='payer_type')
plt.title('Measure Numerator vs Denominator by Payer Type')
plt.xlabel('Measure Denominator')
plt.ylabel('Measure Numerator')
plt.show()
```

# Pie Chart of Assigned Sex at Birth

```python
plt.figure(figsize=(8, 8))
healthcare_df['assigned_sex_at_birth'].value_counts().plot.pie(autopct='%1.1f%%')
plt.title('Distribution of Assigned Sex at Birth')
plt.ylabel('')
plt.show()
```

# Bar Plot of Measure Scaling Factor by County

```python
plt.figure(figsize=(14, 8))
top_counties = healthcare_df['county_name'].value_counts().index[:10]
sns.barplot(data=healthcare_df[healthcare_df['county_name'].isin(top_counties)], x='county_name', y='measure_scaling_factor')
plt.title('Measure Scaling Factor by County (Top 10 Counties)')
plt.xlabel('County Name')
plt.ylabel('Measure Scaling Factor')
plt.xticks(rotation=45)
plt.show()
```

# Box Plot of Measure Denominator by Year

```python
plt.figure(figsize=(12, 6))
sns.boxplot(data=healthcare_df, x='reporting_year', y='measure_denominator')
plt.title('Measure Denominator by Reporting Year')
plt.xlabel('Reporting Year')
plt.ylabel('Measure Denominator')
plt.show()
```

# Violin Plot of Measure Numerator by Age Band

```python
plt.figure(figsize=(12, 6))
sns.violinplot(data=healthcare_df, x='age_band', y='measure_numerator', split=True, hue='assigned_sex_at_birth')
plt.title('Violin Plot of Measure Numerator by Age Band')
plt.xlabel('Age Band')
plt.ylabel('Measure Numerator')
plt.show()
```

# Data Cleaning

# Handling the missing values

```python
healthcare_df['measure_numerator'].fillna(healthcare_df['measure_numerator'].median())
```

```python
print(healthcare_df.isnull().sum())
```

```python
most_frequent_value = healthcare_df['measure_denominator'].mode()[0]
healthcare_df['measure_denominator'].fillna(most_frequent_value)
```

```python
print(healthcare_df.isnull().sum())
```

```python
mean_value = healthcare_df['measure_denominator'].mean()
healthcare_df['measure_denominator'].fillna(mean_value)

median_value = healthcare_df['measure_denominator'].median()
healthcare_df['measure_denominator'].fillna(median_value)
```
